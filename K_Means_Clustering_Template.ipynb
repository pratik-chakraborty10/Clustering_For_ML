# K Means Clustering

## Importing libraries

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Importing dataset

```
dataset=pd.read_csv(" AmazonClusturing_Model.csv")
X=dataset.iloc[:,3:5].values
```

```
print(X)
```

## Optimal number of clusters via Elbow Method

```
from sklearn.cluster import KMeans
wcss=[]

for i in range(1,11):
  kmeans=KMeans(n_clusters=i,init='k-means++',random_state=21)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)

plt.plot(range(1,11),wcss)
plt.title("WCSS VIA ELBOW METHOD")
plt.xlabel("Number of clusters")
plt.ylabel("WCSS value")
plt.show()
```

## K Means Model Training on Training set

```
kmeans=KMeans(n_clusters=4,init='k-means++',random_state=42)#getting the best cluster number via Elbow method
y_means=kmeans.fit_predict(X)
```

```
print(y_means)
```

## Visualizing Clusters

```
#X[y_means==0,0]->means we indicating the income data,X[y_means==0,1]-->we indicating the rating data ---->both are of cluster 0
#X[y_means==1,0]->means we indicating the income data,X[y_means==1,1]-->we indicating the rating data ---->both are of cluster 1
#X[y_means==2,0]->means we indicating the income data,X[y_means==2,1]-->we indicating the rating data ---->both are of cluster 2
#X[y_means==3,0]->means we indicating the income data,X[y_means==3,1]-->we indicating the rating data ---->both are of cluster 3


plt.scatter(X[y_means == 0, 0], X[y_means == 0, 1], s = 100, c = 'magenta', label = 'Cluster 1')#cluster of 0
plt.scatter(X[y_means == 1, 0], X[y_means == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')#cluster of 1
plt.scatter(X[y_means == 2, 0], X[y_means == 2, 1], s = 100, c = 'red', label = 'Cluster 3')#cluster of 2
plt.scatter(X[y_means == 3, 0], X[y_means == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4')#cluster of 3

plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300,c='black',label='Centroids')
plt.title('Cluster of Amazon users')
plt.xlabel('Annual Income in INR')
plt.ylabel('Purchase Rating')
plt.legend()
plt.show()
```
